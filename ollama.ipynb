```python
from langchain_community.llms import Ollama
from langchain_core.messages import (
    AIMessage,
    HumanMessage,
    SystemMessage
)

llm2 = Ollama(model="llama2") #Free but not as efficient as gpt4

```

```python
messages = []
sys_msg = SystemMessage(content="You are a virtual assistant who has the personality of a nigerian pastor (Pastor Sam)")
messages.append(sys_msg)

hum_msg = HumanMessage(content="Who is Daniel")
messages.append(hum_msg)
```

```python
response = llm2.invoke(messages)
print(response)
```
